```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('bds_f_st_release.xlsx', skiprows=9, sheet_name=None)
state_list=[]
for i in df:
    state_list.append(i)

print (state_list)
```

```python
for row in state_list:
    x=df[row]
    x['state']=row
    df[row]=x

cdf = pd.concat(df.values())
```

```python
cdf['start_up_rate']=cdf['Estabs_Entry']/cdf['Emp']
cdf['Growth']=(cdf['Job_Creation']-cdf['Job_Destruction'])/cdf['Emp']
cdf['closure_rate']=cdf['Firmdeath_Estabs']/cdf['Emp']
cdf['vol_rate']=(cdf['Estabs_Entry']+cdf['Firmdeath_Estabs']-abs(cdf['Estabs_Entry']-cdf['Firmdeath_Estabs']))/cdf['Emp']
cdf['turbulance_rate']=(cdf['Estabs_Entry']+cdf['Firmdeath_Estabs'])/cdf['Emp']
cdf['net_entry']=(cdf['Estabs_Entry']-cdf['Firmdeath_Estabs'])/cdf['Emp']
```

```python
cdf.to_csv('states_data.csv')
```

# Get mean values for each decade

```python
# Values for the 80s
headers=['State', "Start_up", "Closure", "Volatility", "Turbulance", "Net_Entry", "Growth"]
rl=[]
for row in state_list:
    regime_list=[]
    df=cdf[cdf['state']==row]
    df=df[df['Year']>1979]
    df=df[df['Year']<1990]
    start_up=df['start_up_rate'].mean()*10
    closure=df['closure_rate'].mean()*10
    vol=df['vol_rate'].mean()*10
    turb=df['turbulance_rate'].mean()*10
    net=df['net_entry'].mean()*10
    growth=df['Growth'].mean()*10
    regime_list=[row, start_up, closure, vol, turb, net, growth]
    rl.append(regime_list)
    
regime=pd.DataFrame(data=rl, columns=headers)
median_start=regime['Start_up'].median()
median_growth=regime['Growth'].median()

regime['s_class']=np.where(regime['Start_up']>median_start,1,-1)
regime['g_class']=np.where(regime['Growth']>median_growth,1,-1)
regime['sum']=regime['s_class']+regime['g_class']

regime['classification']=np.where(regime['sum']==2,"Ent", 0)
regime['classification']=np.where(regime['sum']==0,"Routine", regime['classification'])
regime['classification']=np.where(regime['g_class']==-1,"Rev", regime['classification'])
regime['classification']=np.where(regime['sum']==-2,"Down", regime['classification'])

regime
```

```python
regime.to_csv('regime1980.csv')
```
